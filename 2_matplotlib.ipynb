**Question 1**. Load the data in 'world_bank_turkey.csv' file.

```python
# write your answer here
World_Bank=Table.read_table('world_bank_turkey.csv')
#World_Bank.column_labels
World_Bank=World_Bank.relabeled('Urban population (% of total population)','Urban Ratio').relabeled('Population, total','Total').relabeled('Population, male','Male').relabeled('Population, female','Female').relabeled('Population growth (annual %)','Population Growth').relabeled('Life expectancy at birth, total (years)','Life Expectancy').relabeled('Armed forces personnel, total','Armed Forces Personnel').relabeled('Literacy rate, adult total (% of people ages 15 and above)','Literacy Rate')
```

**Question 3** Add a new column to the table which shows the Rural ratio of the population. Name the column as 'Rural Ratio'

```python
# write your answer here
World_Bank=World_Bank.with_columns('Rural Ratio', World_Bank.column('Rural population (% of total population)')/100)
World_Bank=World_Bank.set_format('Rural Ratio',PercentFormatter)
World_Bank.select('Years','Rural Ratio')
```
