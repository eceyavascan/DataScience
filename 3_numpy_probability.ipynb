**Question 1 (60 Points)**. Several years ago, there was a documentary on Discovery Channel about a Russian girl who claims that she can find the previously injured body parts of the people by just looking at them. The local people were convinced about the girl's talent. It was amazing because she just looks at the people and says "you had an operation or have a problem here" etc. The producer of the document also challenges her talent and he is also shocked that she could find an old injury he had some time ago. So he takes her to US to show scientists and ask them to test her talent. The scientists set up an experiment where there are seven people with known medical history. They write down the medical history of each person and shuffle the cards. Then they ask the girl to match the cards with the correct people. If she can correctly match 5 of them, they will think that she has really some talent in this issue. Note that she cannot have 6 matches because one mistake will also cause another mistake. However, she could correctly match only 4 of them. So they think that her talent is bogus and do not believe her. 
 What is the probability of having 4 matches out of 7 if the cards are distributed randomly? Find the probability by simulation using 10000 trials.

```python
cards=np.arange(1,8)
people=np.arange(1,8)

def talent_show(x,y):
    playing=make_array()
    number_of_a=0
    for i in np.arange(1,x):
        M_C=np.random.choice(cards,size=7,replace=False)
        M_P=np.random.choice(people,size=7,replace=False)
        a=0
        for i in np.arange(0,7):
            if M_C[i]==M_P[i]:
                a=a+1
        playing=np.append(playing,a)    
    number_of_a=np.count_nonzero(playing==y)
    return number_of_a/x
talent_show(10000,4)

#matches=Table().with_column('Number of Matched',playing)
#matches.group('Number of Matched')
```

**Question 2 (10 Points)** Change the number of trials from 1K, 2K, 4K, 8K, 16K, 32K, 64K, and 128K and visualize your results where x axis is the number of trials and y axis is the probability of having 4 matches.

```python
trials=([1000,2000,4000,8000,16000,32000,64000,128000])
probabilities=make_array()

for trial in trials:
    probabilities=np.append(probabilities,talent_show(trial,4))
    
trials_=(['1K', '2K', '4K', '8K', '16K', '32K', '64K', '128K' ])
Results=Table().with_columns('Number of Trials',trials_,'probabilities',probabilities).set_format('probabilities',PercentFormatter)

# write your answer here
```
